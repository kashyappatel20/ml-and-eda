```python
# check null values
df.isnull().sum()
```

```python
# check the null values with the help of list comprehension
[features for features in df.columns if df[features].isnull().sum()>0]
```

```python
# read another csv file 
df_country=pd.read_excel("Country-Code.xlsx",)
```

```python
df_country.head()
```

```python
# combining the columns of country_code dataframe in zomato dataframe columns
final_df=pd.merge(df,df_country,on="Country Code",how="left")
```

```python
final_df.head(2)
```

```python
# check datatype of columns
final_df.dtypes
```

```python
# check which country has higher transaction in graphical manner
```

```python
country_names=final_df.Country.value_counts().index
country_values=final_df.Country.value_counts().values
```

```python
# create the pie chart
plt.pie(country_values[:3],labels=country_names[:3],autopct="%1.2f%%")
```

observation:from the above pie graph we understand that maximum number of transaction are from india after that united state
and then united kingdom

```python
ratings=final_df.groupby(["Aggregate rating","Rating color","Rating text"]).size().reset_index().rename(columns={0:"Rating count"})
```

```python
ratings
```

observation:
1. when rating is between 4.5 to 4.9 it is indicating Excellent ratings
2. when rating is between 4.0 to 4.4 it is indicating very good ratings
3. when rating is between 3.5 to 4.9 it is indicating good ratings
4. when rating is between 2.5 to 3.4 it is indicating average ratings 
5. when rating is between 1.8 to 2.4 it is indicating poor ratings
6. when rating was 0,0 it is indicating no ratings 

```python
plt.figure(figsize=(12,7))
sns.barplot(x="Aggregate rating",y="Rating count",data=ratings)
```

```python
plt.figure(figsize=(12,7))
sns.barplot(x="Aggregate rating",y="Rating count",data=ratings,hue="Rating color",palette=["white","red","orange","yellow","green","green"])
```

observation:
1. not rated count is very high
2. Maximum ratings are between 2.5 to 3.4 

```python
# plot data using countplot
sns.countplot(x="Rating color",data=ratings,palette=["grey","red","orange","yellow","green","green"])
```
